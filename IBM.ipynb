```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from imblearn.over_sampling import SMOTE

# 1. قراءة البيانات
data = pd.read_csv('IBM_HR.csv')

# 2. Encoding
data_encoded = data.copy()
str_data = []
label_encoders = {}

for i in data.columns:
    if data[i].dtype == 'object':
        str_data.append(i)
        le = LabelEncoder()
        data_encoded[i] = le.fit_transform(data[i])
        label_encoders[i] = le

# 3. تحديد X_cols (✅ الصح)
X_cols = [col for col in data_encoded.columns if col != 'Attrition']

# حذف أعمدة غير مفيدة
cols_to_remove = ['EmployeeCount', 'Over18', 'StandardHours']
X_cols = [col for col in X_cols if col not in cols_to_remove]

# 4. SMOTE
oversample = SMOTE()
x, y = oversample.fit_resample(data_encoded[X_cols], data_encoded['Attrition'])

# 5. دمج
final_data = pd.DataFrame(x, columns=X_cols)
final_data['Attrition'] = y

# 6. إرجاع encoding
for i in str_data:
    if i in final_data.columns:
        final_data[i] = label_encoders[i].inverse_transform(
            final_data[i].round().astype(int)
        )


```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
scaler= StandardScaler().fit(x_train)
normalized_x_train= scaler.transform(x_train)
normalized_x_test= scaler.transform(x_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV


param_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'max_features': ['sqrt', 'log2']
}
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(random_state=42),
    param_grid=param_grid,
    cv=5,
    scoring='accuracy',
    n_jobs=-1,
    verbose=2
)

grid_search.fit(normalized_x_train, y_train)

print(f"Best Parameters: {grid_search.best_params_}")
print(f"Best Score: {grid_search.best_score_:.4f}")

best_model = grid_search.best_estimator_
y_pred = best_model.predict(normalized_x_test)
print(f"Test Score: {best_model.score(normalized_x_test, y_test):.4f}")
```

```python
from sklearn.ensemble import RandomForestRegressor
model_rf = RandomForestClassifier (
    max_depth=30,
    max_features='log2',
    min_samples_leaf=1,
    min_samples_split=2,
    n_estimators=300,
    random_state=42
)
model_rf.fit(normalized_x_train,y_train)
y_pred=model_rf.predict(normalized_x_test)
model_rf.score(normalized_x_test,y_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```
